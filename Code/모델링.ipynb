# 모델링

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
x_data = predict.drop(['영업상태구분코드'],axis=1)
y_data = predict['영업상태구분코드']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=777, stratify= y_data)

print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(criterion='squared_error')
model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor(criterion = 'squared_error', random_state=5000)

params = {'n_estimators': [350, 450, 550],
          'max_features': [4, 5, 6],
          'min_samples_leaf': [1, 2.5, 4.5]}

greedy_CV = GridSearchCV(model, param_grid=params, cv = 3, n_jobs = -1)
greedy_CV.fit(x_train, y_train)
```

```python
pred = greedy_CV.predict(x_test)
```

```python
pred
```

```python
score = np.round(pred,0)
score = pd.DataFrame(score)
```

```python
score
```

```python
from sklearn.metrics import accuracy_score
print('정확도는 : {0: .4f}'.format(accuracy_score(y_test,score)))
```
